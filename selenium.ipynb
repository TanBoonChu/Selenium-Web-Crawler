Selenium Web Crawler

To develop web crawler to extract information from dynamic webpages using Selenium.
Objectives: To generate a list of grants available for startups in Malaysia. 
Useful Selenium Manual: https://selenium-python.readthedocs.io/

```python
#import libraries
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support.ui import Select
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
```

```python
#Create a list 
Grants = []
#target dynamic website
address = "https://muru-ku.com/grants"
#locate Chrome driver in 
driver = webdriver.Chrome(service=Service(r"C:\Users\ASUS VivoBook\Downloads\chromedriver_win32\chromedriver.exe"))
#Read URL
driver.implicitly_wait(10)
driver.get(address)
#Press the pop out window button
elements = driver.find_elements(By.XPATH, '//button[text()="Read on"]')

#Create iterative actions to click the pop out windows, extract details, and close the windows.
for btn in elements:
    btn.click()
    ele = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, '/html/body/div[3]/div/div/div[1]/h1')))
    grant = ele.text
    ele = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, '/html/body/div[3]/div/div/div[1]/div[4]')))
    description = ele.text
    ele = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, '/html/body/div[3]/div/div/div[1]/div[5]')))
    criteria = ele.text
    ele = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, '/html/body/div[3]/div/div/div[1]/button')))
    ele.click()
#To extract Javascript interative pop out tab link
    window = driver.current_window_handle
    pop = driver.window_handles[1]
    driver.switch_to.window(pop)
    link = driver.current_url
    driver.close()
    driver.switch_to.window(window)
#close the pop out window
    element = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH,'//button[text()="Close"]')))
    element.click()
#Append info into dictionary
    Grants.append({
        'Grant':grant,
        'Description':description[13:],
        'Criteria':criteria[14:],
        'Link':link
    })
print(Grants)
driver.close()
#Sit back and relax while Selenium is scraping the web for you =D
```

```python
#To save as Excel sheet
df = pd.DataFrame(data=Grants)
writer = pd.ExcelWriter('Grant.xlsx', engine='xlsxwriter')
df.to_excel(writer, sheet_name='Malaysia', index=False)
writer.save()
```

To crawl data from webpages having dropdown list.

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support.ui import Select
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

address = "https://mymetdata.met.gov.my/shop/product/9/records-of-daily-rainfall-amount-08-00-08-00-mst"
driver = webdriver.Chrome(service=Service(r"C:\Users\ASUS VivoBook\Downloads\chromedriver_win32\chromedriver.exe"))
driver.get(address)
select = Select(driver.find_element("id","listState"))
print([i.text for i in select.options])
select.select_by_value("1")
element = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.ID, "listDistrict")))
select = Select(element)
print([i.text for i in select.options])
select.select_by_index(1)
element = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.ID, "listStation")))
select = Select(element)
print([i.text for i in select.options])
select.select_by_index(1)
element = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH,'//*[@id="link-show-map"]/a')))
element.click()
element = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH,'//*[@id="show-map-elevation-val"]')))
text = element.text
print(text)
driver.close()
```
